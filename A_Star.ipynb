```python
#Nguyễn Chí Nguyên - 2001240306
#Bài 2
from collections import deque

class Graph:
    def __init__(self, adjac_lis):
        self.adjac_lis = adjac_lis
    def get_neighbors(self, v):
        return self.adjac_lis[v]
    def h(self, n):
        H = {'A': 1, 'B': 1, 'C': 1, 'D': 1}
        return H[n]

    def a_star_algorithm(self, start, stop):
        open_lst = set([start])
        closed_lst = set([])
        g = {start: 0}
        par = {start: start}

        step = 0
        while open_lst:
            step += 1
            # Tìm node có f(n) thấp nhất
            n = min(open_lst, key=lambda v: g[v] + self.h(v))

            print(f"--- Bước {step} ---")
            print(f"Node đang xét: {n}")
            print(f"Open list: {open_lst}")
            print(f"Closed list: {closed_lst}")
            print(f"f({n}) = g({n}) + h({n}) = {g[n]} + {self.h(n)} = {g[n] + self.h(n)}")
            print("-" * 20)

            if n == stop:
                path = []
                while par[n] != n:
                    path.append(n)
                    n = par[n]
                path.append(start)
                path.reverse()
                return path

            for (m, weight) in self.get_neighbors(n):
                if m not in open_lst and m not in closed_lst:
                    open_lst.add(m)
                    par[m] = n
                    g[m] = g[n] + weight
                elif g[m] > g[n] + weight:
                    g[m] = g[n] + weight
                    par[m] = n
                    if m in closed_lst:
                        closed_lst.remove(m)
                        open_lst.add(m)

            open_lst.remove(n)
            closed_lst.add(n)

        return None

adjac_lis = {'A': [('B', 1), ('C', 3), ('D', 7)], 'B': [('C', 1), ('D', 5)], 'C': [('D', 12)]}
graph = Graph(adjac_lis)
result_path = graph.a_star_algorithm('A', 'D')
print("\n=> Kết quả cuối cùng:", result_path)
```

```python
#Bài Tập Về Nhà
#Bài 2
import heapq

class TSP_AStar:
    def __init__(self, matrix, cities):
        self.matrix = matrix  # Ma trận kề chi phí giữa các thành phố
        self.cities = cities  # Danh sách tên các thành phố
        self.n = len(cities)

    def h(self, current_city_idx, unvisited_mask):
        """
        Hàm Heuristic: Ở đây dùng chi phí cạnh nhỏ nhất
        để nối các thành phố còn lại.
        """
        if not unvisited_mask:
            # Nếu đã đi hết, h(n) là chi phí quay về điểm xuất phát (0)
            return self.matrix[current_city_idx][0]

        # Đơn giản hóa: h(n) = khoảng cách đến thành phố chưa đi gần nhất
        min_dist = float('inf')
        for i in range(self.n):
            if (unvisited_mask >> i) & 1:
                min_dist = min(min_dist, self.matrix[current_city_idx][i])
        return min_dist

    def solve(self):
        # Trạng thái: (f_score, g_score, current_city_idx, visited_mask, path)
        # visited_mask dùng bitmask để tối ưu tốc độ và bộ nhớ
        # chỗ này là f , g ,curr_city ,visited_city, và list lưu lại hành trình
        start_node = (0, 0, 0, 1 << 0, ['A'])

        open_lst = [start_node]

        step = 0
        while open_lst:
            f, g, u, mask, path = heapq.heappop(open_lst)
            step += 1

            print(f"--- Bước {step} ---")
            print(f"Đang tại: {self.cities[u]} | Lộ trình: {' -> '.join(path)}")
            print(f"g(n) = {g}, h(n) = {f-g} => f(n) = {f}")

            # Nếu đã đi qua tất cả các thành phố
            if mask == (1 << self.n) - 1:
                total_g = g + self.matrix[u][0]
                final_path = path + [self.cities[0]]
                print("-" * 20)
                print(f"==> Đã hoàn thành chu trình!")
                return final_path, total_g

            # Khám phá các thành phố lân cận
            for v in range(self.n):
                if not ((mask >> v) & 1): # Nếu thành phố v chưa ghé thăm
                    new_g = g + self.matrix[u][v]
                    new_mask = mask | (1 << v)
                    new_h = self.h(v, new_mask)
                    new_f = new_g + new_h

                    heapq.heappush(open_lst, (new_f, new_g, v, new_mask, path + [self.cities[v]]))

            print("-" * 20)

        return None, float('inf')

# --- Thiết lập dữ liệu ---
# Giả sử ta có 4 thành phố A, B, C, D
cities = ['A', 'B', 'C', 'D']
# Ma trận chi phí (0 là đường đi đến chính nó)
matrix = [
    [0, 10, 15, 20],
    [10, 0, 35, 25],
    [15, 35, 0, 30],
    [20, 25, 30, 0]
]

tsp_solver = TSP_AStar(matrix, cities)
path, total_cost = tsp_solver.solve()

print(f"\n=> Lộ trình tối ưu: {' -> '.join(path)}")
print(f"=> Tổng chi phí: {total_cost}")
```
