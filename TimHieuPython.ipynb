```python
#Bai 14d
def tinhtongdongk(k,matrix):
  tong = 0
  for i in range(len(matrix[k])):
    tong+=matrix[k][i]
  return tong


matrix = [[5, 9, 1, 6], [9, 7, 2, 10], [6, 8, 3, 3]]
max = tinhtongdongk(0,matrix)
index = 0
for i in range(1,len(matrix)):
  if(max < tinhtongdongk(i,matrix) ):
    max = tinhtongdongk(i,matrix)
    index = i
print(f"index = {index} , có tổng bằng {max}")
```

```python
#Bai 14e

def tinhtichcotk(k,matrix):
  tich = 1
  for j in range(len(matrix)):
    tich*=matrix[j][k]
  return tich
matrix = [[5, 9, 1, 6], [9, 7, 2, 10], [6, 8, 3, 3]]

max = tinhtichcotk(0,matrix)
index = 0
for j in range(len(matrix)):
  if(max < tinhtichcotk(j,matrix) ):
    index = j
    max = tinhtichcotk(j,matrix)
print(f"cột có tích lớn nhất là cột {index} tích bằng {max} ")

```

```python
#Bai2a
matrix = [[5, 9, 1, 6],
          [9, 7, 2, 10],
          [6, 8, 3, 3],
          [-2, 4, 0, 5]]

def tinhTongTamGiacTrenDCP(matrix):
  tong = 0
  for i in range(len(matrix)):
    for j in range(len(matrix[i])):
      if(i+j<len(matrix)-1):
        tong+=matrix[i][j]
  return tong

print(f"tổng các phần tử thuộc tam giác trên của ma trận là: {tinhTongTamGiacTrenDCP(matrix)}")
```

```python
#Bai2b
def chuyenAmThanhABS(matrix):
  for i in range(len(matrix)):
    for j in range(len(matrix[i])):
      if(matrix[i][j]<0):
        matrix[i][j] = abs(matrix[i][j])
  return matrix

print("ma trận sau khi chuyển các phần tử âm thành trị tuyệt đối là : \n")
for row in chuyenAmThanhABS(matrix):
    for item in row:
        print(item, end=" ")
    print("\n", end="")
```

```python
#bai2e
def kTraMatrixDoiXung(matrix):
  for i in range(len(matrix)):
    for j in range(len(matrix[i])):
      if(matrix[i][j]!=matrix[j][i]):
        return False
  return True
if(kTraMatrixDoiXung(matrix)):
  print("ma trận là ma trận đối xứng!")
else:
  print("ma trận không đối xứng!")
```

```python
#Bai2f
def kTraDCCASC(matrix):
    n = len(matrix)
    if n == 0: return True
    for i in range(1, n):
        if matrix[i][i] < matrix[i-1][i-1]:
            return False
    return True

if kTraDCCASC(matrix):
    print("Đường chéo chính tăng dần!")
else:
    print("Đường chéo chính không tăng dần!")
```

```python
#Bai2h
matrix = [[5, 9, 1, 6],
          [9, 7, 2, 10],
          [6, 8, 3, 3],
          [-2, 4, 0, 5]]

def kTraDCPASC(matrix):
    n = len(matrix)
    if n <= 1: return True

    for i in range(1, n):
        current = matrix[i][n - 1 - i]
        previous = matrix[i - 1][n - i]
        if current < previous:
            return False
    return True

if kTraDCPASC(matrix):
    print("Đường chéo phụ tăng dần!")
else:
    print("Đường chéo phụ không tăng dần!")
```
